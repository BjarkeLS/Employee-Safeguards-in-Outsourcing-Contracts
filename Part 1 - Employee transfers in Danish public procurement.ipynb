```python
corpus = pd.concat(frames)
```

## Data cleaning

```python
def clean(text):
    text = re.sub('\s\s+', ' ', str(text)) # add spaces at line breaks
    text = re.sub("\n", "", text) # remove line breaks
    text = re.sub("\r", " ", text) # remove line breaks in doc
    text = re.sub("[()]","", text) # remove parentheses
    text = re.sub("%", "", text) # remove percentage
    text = re.sub('[-()\"#/@;:<>{}`+=~|.!?,‰]', ' ', text) # remove special characters and symbols
    text = re.sub(' +', ' ', text) # remove too many spaces
    text = text.lower()
    return text
```

```python
corpus['clean_text'] = corpus.text.apply(lambda x : clean(x))
```

```python
corpus.head(30)
```

## Word Tokenization

```python
corpus['word'] = corpus.clean_text.apply(lambda x : word_tokenize(x))
corpus.head()
```

## Lemmatization

```python
lemmatizer = lemmy.load("da")
```

```python
def lemmy_definition(words):
    words = [lemmatizer.lemmatize("", w)[0] for w in words]
    return words
```

```python
corpus['lemmy'] = corpus.word.apply(lambda x : lemmy_definition(x))
```

```python
corpus.head()
```

## Filter employee transfers

```python
virksomhedsoverdragelse = ['virksomhedsoverdragelse']
```

```python
def rule(lemmy, virksomhedsoverdragelse):
    return [n for n in lemmy if
             any(m in n for m in virksomhedsoverdragelse)]
```

```python
corpus['virksomhedsoverdragelse'] = corpus.lemmy.apply(lambda x, y=virksomhedsoverdragelse : rule(x, y))
```
